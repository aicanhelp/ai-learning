```java
import java.util.List;
import java.util.ArrayList;
import java.util.Arrays;

public class FindNumbersWithoutLen{
    private int max;
    public FindNumbersWithoutLen(int max){
        this.max=max;
    }
    
    public List<Integer> find(int[] values){
        return findSub(values,0,this.max);
    }

    private List<Integer> findSub(int[] values, int fromIndex, int remainder){
        List<Integer> subs=new ArrayList<>();
        
        if (values[fromIndex]==remainder){
            subs.add(values[fromIndex]);
            return subs;
        }
        
        if(fromIndex<values.length-1 && remainder>values[fromIndex]){
            List<Integer> nextSubs=findSub(values,fromIndex+1, remainder-values[fromIndex]);
            if (!nextSubs.isEmpty()){
                subs.add(values[fromIndex]);
                subs.addAll(nextSubs);
                return subs;
            }else
                return findSub(values, fromIndex+1, remainder);
        }
        
        return subs;
    }
}

public class FindNumbersWithLen{
    private int max,len;
    public FindNumbersWithLen(int max, int len){
        this.max=max;
        this.len=len;
    }
    
    public List<Integer> find(int[] values){
        return findSub(values,0,0,this.max);
    }

    private List<Integer> findSub(int[] values, int foundLen, int fromIndex, int remainder){
        List<Integer> subs=new ArrayList<>();
        
        if (values[fromIndex]==remainder){
            if (foundLen+1==this.len)
                subs.add(values[fromIndex]);
            return subs;
        }
        
        if(fromIndex<values.length-1 && remainder>values[fromIndex]){
            List<Integer> nextSubs=findSub(values,foundLen+1, fromIndex+1, remainder-values[fromIndex]);
            if (!nextSubs.isEmpty()){
                subs.add(values[fromIndex]);
                subs.addAll(nextSubs);
                return subs;
            }else
                return findSub(values,foundLen, fromIndex+1, remainder);
        }
        
        return subs;
    }
}
```
